```python
import numpy as np
import matplotlib.pyplot as plt  
x1=[500000 , 1000000, 2000000, 3000000]
y1=[50.7861, 100.456, 208.648,  309.213]


l1=plt.plot(x1,y1,'r--')

plt.plot(x1,y1,'ro-')
plt.title('Single Thread Version K-means Performance With Different Number of Data Points')
plt.xlabel('Number of Data Points')
plt.ylabel('Running Time in Seconds)')
plt.legend()
plt.show()
```

```python
x1=np.array([1 , 2, 4,8, 16])
y1=np.array([309.213, 197.973, 151.426,   111.441, 95.4495])
y2 = y1 / 309.213

l1=plt.plot(x1,y1,'r--')

plt.plot(x1,y1,'ro-')
plt.title('MPI Version K-means Performance With Different Cores')
plt.xlabel('Number of Cores')
plt.ylabel('Running Time in Seconds)')
plt.legend()
plt.show()
```

```python
x1=np.array([1 , 2, 4,8, 16])
y1=np.array([309.213, 197.973, 151.426,   111.441, 95.4495])
y2 = []
for i in range(5):
    y2.append(y1[0] /  (y1[i] * (i + 1)))

l1=plt.plot(x1,y2,'g--')

plt.plot(x1,y2,'go-')
plt.title('Speedup Ratio for Different Cores')
plt.xlabel('Number of Cores')
plt.ylabel('Ratio')
plt.legend()
plt.show()
```
